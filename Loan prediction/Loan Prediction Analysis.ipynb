```python
df.isnull().sum()
```

```python
#numerical attributes(mean)
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
#categorical attributes (mode)
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
# drop unnecessary columns
cols = ['ApplicantIncome', 'CoapplicantIncome', "LoanAmount", "Loan_Amount_Term", "Total_Income", 'Loan_ID', 'CoapplicantIncomeLog']
df = df.drop(columns=cols, axis=1)
df.head()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Gender',"Married","Education",'Self_Employed',"Property_Area","Loan_Status","Dependents"]
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

**Train-Test Split**

```python
# specify input and output attributes
X = df.drop(columns=['Loan_Status'], axis=1)
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

**Model Training**

```python
#import cross_val_score function
from sklearn.model_selection import cross_val_score
#function to classify model
def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
    #train data using training data
    model.fit(x_train, y_train)
    #print accuracy
    print("Accuracy is", model.score(x_test, y_test)*100)
    # cross validation - used for better validation of the model
    # eg: cv-5 , train-4, test-1 
    #dataset is divided into 5 folds, model is trained and evaluated 5 times
    score = cross_val_score(model, x, y, cv=5)
    #mean accuracy across all folds 
    print("Cross validation is", np.mean(score)*100)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model, X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model, X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
model = RandomForestClassifier()
classify(model, X, y)
```

```python
model = ExtraTreesClassifier()
classify(model, X, y)
```

**Hyperparameter Tuning**

```python
model = RandomForestClassifier(n_estimators=150, min_samples_split=25, max_depth=7, max_features=1)
classify(model, X, y)
```

**Confusion Matrix**

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print("confusion matrix is \n", cm)
```
